### Question 1


which issue should i solve first ?? and why

You should solve the Gaussian noise issue first before applying contrast enhancement. Here's why:

Noise Removal: Gaussian noise can obscure important details in the image. Removing noise first ensures that the subsequent contrast enhancement is applied to a cleaner image, which can improve the effectiveness of the contrast enhancement.

Contrast Enhancement: Enhancing contrast on a noisy image can amplify the noise, making it more prominent and potentially degrading the image quality. By removing the noise first, you ensure that the contrast enhancement improves the visibility of actual image features rather than the noise.

```python
''' Question 1'''

# Read the image and load section
img_rgb = preprocess_image('Q1.png', 'rgb')

# Solving Issues section

# ISSUES #1 -> Gaussian Noise
img_rgb_c1 = remove_noise(img_rgb, 'gaussian', sigma=2)

# ISSUES #2 -> Contrast Enhancement
img_rgb_c2 = enhance_contrast(img_rgb_c1, method='clahe', clip_limit=3.0, tile_grid_size=(8, 8))

# Example usage:
# img_fixed = adjust_exposure(img, method='clahe', clip_limit=3.0)
# img_fixed = correct_perspective(img, [[x1,y1], [x2,y2], [x3,y3], [x4,y4]])
# img_fixed = correct_color_cast(img, method='white_balance', percentile=95)
# img_fixed = reduce_noise(img, method='nlmeans', h=10)

# show images and outputs section
show_images([img_rgb,img_rgb_c1,img_rgb_c2], ['original','after noise removal','after contrast enhancement'])
```

```python
''' Question 2'''

# 1. Initial preprocessing and noise removal
img_rgb = preprocess_image('Q1.png', 'rgb')
img_rgb_c1 = remove_noise(img_rgb, 'gaussian', sigma=2)

# 2. Enhance contrast using CLAHE
img_rgb_c2 = enhance_contrast(img_rgb_c1, method='clahe', clip_limit=3.0, tile_grid_size=(8, 8))

# 3. Correct color cast (yellow tint)
img_rgb_c3 = correct_color_cast(img_rgb_c2, method='white_balance', percentile=95)

# 4. Final gentle denoising if needed
img_rgb_final = reduce_noise(img_rgb_c3, method='nlmeans', h=5)

# Show all steps
show_images([
    img_rgb,
    img_rgb_c1,
    img_rgb_c2,
    img_rgb_c3,
    img_rgb_final
], [
    'Original',
    'After noise removal',
    'After contrast enhancement',
    'After color correction',
    'Final result'
])
```
